## 2. Estimación de Parámetros por MCO

```python
def generar_resultados_lin_reg(X,y):
    reg = LinearRegression().fit(X, y)
    score = reg.score(X, y)
    print('Score: '+ str(score))
    X_columns = list(X.columns)
    coef = list(reg.coef_)
    coef_zip = zip(X_columns,coef)
    coef_dict = dict(coef_zip)
    coeficientes = pd.DataFrame(coef_dict,index = ['Coeficientes'])
    print(coeficientes)
    y_predict = pd.DataFrame(reg.predict(X))
    y_predict.rename({0:'Predicción'},axis=1,inplace=True)
    y = pd.DataFrame(y).reset_index(drop=True)
    y.rename({'lpsa':'Target'},axis=1,inplace=True)
    tabla = pd.concat([y,y_predict],axis = 1)
    tabla['Errores'] = tabla['Target'] - tabla['Predicción']
    ECM = np.mean(tabla['Errores']**2)
    RECM = np.sqrt(ECM)
    MAE = np.mean(abs(tabla['Errores']))
    MAPE = np.mean(abs(tabla['Errores']/tabla['Target']))
    datos = {'ECM':ECM,'RECM':RECM,'MAE':MAE,'MAPE':MAPE}
    eval = pd.DataFrame(datos, index = ['Valores'])
    print(eval)
    return tabla
```

```python
generar_resultados_lin_reg(data_entrenamiento_1.iloc[:,:8],data_entrenamiento_1['lpsa'])
```

```python
generar_resultados_lin_reg(data_testing_1.iloc[:,:8],data_testing_1['lpsa'])
```

```python
def data_estandarizada(data):
    data_estandarizada = pd.DataFrame()
    for col in list(data.columns):
        data_estandarizada[col] = (data[col] - data[col].mean())/(data[col].std())
    return data_estandarizada
```

```python
data_train_stand = data_estandarizada(data_entrenamiento_1)
data_test_stand = data_estandarizada(data_testing_1)
```

```python
generar_resultados_lin_reg(data_train_stand.iloc[:,:8],data_train_stand['lpsa'])
```

```python
generar_resultados_lin_reg(data_test_stand.iloc[:,:8],data_test_stand['lpsa'])
```
